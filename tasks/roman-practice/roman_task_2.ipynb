```python
df_test = pd.read_csv('/content/test.csv')
```

```python
import sklearn
```

```python
model = sklearn.cluster.KMeans(n_clusters=4)
model.fit(df_train.drop(['SampleID'], axis=1))
```

```python
sub = pd.DataFrame({'SampleID': df_test['SampleID'], 'Label': model.predict(df_test.drop(['SampleID'], axis=1))})
sub.to_csv('sub_for_roman.csv', index=False)
```
